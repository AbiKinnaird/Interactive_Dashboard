## Random Forest implementation

```python
def RandomForest(dataframe, num_samples):
    """
    This function implements the Random Forest model
    
    :param dataframe: Credit card fraud data
    :param num_samples: number of samples to train the model on
    :return: outputs a dictionary of the random forest model and the trained X and Y data
    
    """
    #Preprocessing into the test and training sets
    df = dataframe.sample(n=num_samples, random_state=42)
    y = df['Class']
    X = df.drop('Class', axis = 1)
    X_train, X_test, Y_train, Y_test = train_test_split(X, y, random_state=0)

    #Apply SMOTE to even the dataset
    smote = SMOTE(random_state=99)
    X_train_resampled, Y_train_resampled = smote.fit_resample(X_train, Y_train)

    #Train the model
    rf = RandomForestClassifier(n_estimators=100, max_depth = 5,
                                random_state=99)
    rf.fit(X_train_resampled, Y_train_resampled)
    
    return {'df':df,'rf':rf,'X_train':X_train,'Y_train':Y_train, 'X_resampled':X_train_resampled, 'Y_resampled':Y_train_resampled, 'X_test':X_test,'Y_test':Y_test}
```

```python
#Set up the variables 
RandomForest = RandomForest(df,150000)
df = RandomForest['df']
rf = RandomForest['rf']
X_train = RandomForest['X_train']
Y_train = RandomForest['Y_train']
X_test = RandomForest['X_test']
Y_test = RandomForest['Y_test']
X_train_resampled = RandomForest['X_resampled']
Y_train_resampled = RandomForest['Y_resampled']
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn import metrics

#Create a plot to see the effect of changing the max depth

fig, axs = plt.subplots(figsize=(10, 3))
Max = range(4,12)
accuracy = np.zeros(len(Max))
auc = np.zeros(len(Max))

#loop over many differnet max depths and calculate accuracy
for i, max_depth in enumerate(Max):
    rf_clf = RandomForestClassifier(n_estimators=100, max_depth=max_depth, n_jobs=-1)
    rf_clf.fit(X_train_resampled, Y_train_resampled)
    y_pred = rf_clf.predict(X_test)
    accuracy[i] = accuracy_score(Y_test, y_pred)
    y_pred_proba = rf_clf.predict_proba(X_test)
    auc[i] = metrics.roc_auc_score(Y_test, y_pred_proba[:,1])


axs.plot(Max, accuracy, label = 'Accuracy', marker = 's')
axs.plot(Max, auc, label = 'AUC', marker = 's')

axs.set_title("Model Performance Metrics")
axs.set_xlabel("Maximum Depth")
axs.set_ylabel("Score")

axs.legend()
plt.show()
```

## Clustering

```python
from sklearn.metrics import accuracy_score
from Levenshtein import distance as lev

class ClusteringData:
    """
    This class constructs the data for clustering of the trees. Involving the construction a string similarity
    comparison and an accuracy metric measure. 
    """
    
    def __init__(self, RandomForest, overall_children, X_train, Y_train, X_test, Y_test):
        self.m_RandomForest = RandomForest
        self.m_overall_children = overall_children
        self.m_X_train = X_train
        self.m_Y_train = Y_train
        self.m_X_test = X_test
        self.m_Y_test = Y_test
    
    def DT_Accuracy(self):
        """
        Find the accuracy for each Decision Tree and append to a list for plotting
        """
        j = 0
        accuracys = []
        for j in range(100):
            self.m_RandomForest[j].fit(self.m_X_train, self.m_Y_train)
            y_pred = self.m_RandomForest[j].predict(self.m_X_test)
            accuracys.append(accuracy_score(self.m_Y_test, y_pred))
        return accuracys
   
    def Score_for_level(self, concat_list):
        """
        Create a summed score of how differenet each list (i.e The decision tree structure) is 
        from each other list.
        """
        
        level = []
        for j in range(len(concat_list)):
            list_score =[]
            for n in range(len(concat_list)):
                if j != n:
                    list_score.append(lev(concat_list[j], concat_list[n]))
            list_sum = sum(list_score)
            level.append(list_sum)

        return level
    
    def Fix_string(self, child_list):
        """
        Takes in a string from the output of a heirachy list which was constructed to have small differences 
        to allow construction of sunburst/treemap plot but to appear as the same value in the visual.
        """
        
        child_list = ['None'  if item == None else item for item in child_list]

        for i in range(1,24):
            target = 'Leaf Node' + ' '*i
            child_list = ['Leaf Node'  if item == target else item for item in child_list]

        child_list = ['Leaf node'  if item == '  Leaf Node  ' else item for item in child_list]
        child_list = ['Leaf node'  if item == '   Leaf Node    ' else item for item in child_list]
        child_list = ['Leaf node'  if item == '    Leaf Node    ' else item for item in child_list]
        child_list = ['Leaf node'  if item == 'Leaf Node    ' else item for item in child_list]
        child_list = ['Leaf node'  if item == ' Leaf Node  ' else item for item in child_list]
        return child_list
    
    def Overall_score(self):
        """
        Finds the overall disimilarity score
        """
        
        level_1 = []
        for n in range(100):
            child_list = self.m_overall_children[n]
            concat = ''.join(self.Fix_string(child_list[0:2]))
            level_1.append(concat)

        level_1_score = np.array(self.Score_for_level(level_1))*4

        level_2 =[]
        for n in range(100):
            child_list = self.m_overall_children[n]
            concat = ''.join(self.Fix_string(child_list[2:6]))
            level_2.append(concat)

        level_2_score = np.array(self.Score_for_level(level_2))*3

        level_3 =[]
        for n in range(100):
            child_list = self.m_overall_children[n]
            concat = ''.join(self.Fix_string(child_list[6:14]))
            level_3.append(concat)

        level_3_score = np.array(self.Score_for_level(level_3))*2

        level_4 =[]
        for n in range(100):
            child_list = self.m_overall_children[n]
            concat = ''.join(self.Fix_string(child_list[14:30]))
            level_4.append(concat)

        level_4_score = np.array(self.Score_for_level(level_4))

        #Find the combined score for every level
        Total_score = level_1_score + level_2_score + level_3_score + level_4_score 
        return Total_score
   
```

## Summary Treemap

```python
class SummaryTreemapProcessor:
    """
    Produce a large hierachy structure that only has the unique values, construct this in a dataframe with columns
    'Random Forest', 'Level','Value'
    """
    
    def __init__(self, level_new, overall_children_cluster):
        self.m_level_new = level_new
        self.m_overall_children_cluster = overall_children_cluster
        self.m_values_unique = []
        self.m_level_treemap = []
        self.m_constant = []
        self.m_unique_0 = []
        self.m_unique_1 = []
        self.m_unique_2 = []
        self.m_unique_3 = []
        self.m_unique_4 = []
        self.m_size_0 = 0
        self.m_size_1 = 0
        self.m_size_2 = 0
        self.m_size_3 = 0
        self.m_size_4 = 0
        self.m_size = []

    
    def RoundingString(self, list_of_strings):
        """
        Function for rounding the output of the list with 0.1 values to make them all rounded
        """
        rounded_list = []
        for item in list_of_strings:
            item_new = float(item[1:])
            item_new = round(item_new,0)
            rounded_list.append('V' + str(item_new))
        return rounded_list

    def ChildList(self):
        """
        Function for making the list of all the unique values for each of the levels
        """
        
        df_child = pd.DataFrame(columns = self.m_level_new)
        for child in self.m_overall_children_cluster:
            df_child.loc[len(df_child)] = child
        df_child

        unique_0 = df_child[0].unique()
        self.m_size_0 = df_child[0]
        unique_0 = list(unique_0)

        unique_11 = df_child[1.1].unique() 
        unique_12 = df_child[1.2].unique()
        unique_1 = list(set(list(unique_11) + list(unique_12)))
        unique_1 = self.RoundingString(unique_1)

        unique_1 = ['V10.0'  if item == 'V10.1' else item for item in unique_1]
        unique_1 = ['V11.0'  if item == 'V11.1' else item for item in unique_1]
        self.m_size_1 = self.RoundingString(list(df_child[1.1]) + list(df_child[1.2]))
        unique_1 = set(unique_1)

        unique_21 = df_child[2.1].unique() 
        unique_22 = df_child[2.2].unique()
        unique_23 = df_child[2.3].unique() 
        unique_24 = df_child[2.4].unique()
        unique_2 = list(set(list(unique_21) + list(unique_22)+list(unique_23)+list(unique_24)))

        unique_2 = set(unique_2)

        for i in range(0,24):
            target = 'VLeaf node' + ' '*i
            unique_2 = ['V30'  if item == target else item for item in unique_2]
        
        self.m_size_2 = list(list(df_child[2.1]) + list(df_child[2.2])+ list(df_child[2.3]) + list(df_child[2.4]))
        
        for i in range(0,24):
            target = 'VLeaf node' + ' '*i
            self.m_size_2 = ['V30'  if item == target else item for item in self.m_size_2]
        
        self.m_size_2 = self.RoundingString(self.m_size_2)
        
        unique_2 = set(self.RoundingString(unique_2))

        unique_3 = []

        list_3 = level_new[7:14]
        for i in list_3:
            unique_3 = list(unique_3) + list(df_child[i].unique())
        
        for i in list_3:
            self.m_size_3 = list(unique_3) + list(df_child[i])
        
        
        unique_3 = set(unique_3)

        for i in range(0,24):
            target = 'VLeaf node' + ' '*i
            unique_3 = ['V30'  if item == target else item for item in unique_3]
        
        for i in range(0,24):
            target = 'VLeaf node' + ' '*i
            self.m_size_3 = ['V30'  if item == target else item for item in self.m_size_3]
        
        
        unique_3 = set(self.RoundingString(unique_3))
        self.m_size_3 = self.RoundingString(self.m_size_3)

        list_4 = level_new[15:]
        unique_4 = []

        for i in list_4:
            unique_4 = list(unique_4) + list(df_child[i].unique())
            
        for i in list_4:
            self.m_size_4 = list(unique_4) + list(df_child[i])

        for i in range(1,24):
            target = 'Leaf Node' + ' '*i
            unique_4 = ['Leaf Node'  if item == target else item for item in unique_4]

        for i in range(0,24):
            target = 'VLeaf node' + ' '*i
            self.m_size_4 = ['V30'  if item == target else item for item in self.m_size_4]   
            
            
        unique_4 = set(unique_4)   

        for i in range(0,24):
            target = 'VLeaf node' + ' '*i
            unique_4 = ['V30'  if item == target else item for item in unique_4]
        
        self.m_size_4 = self.RoundingString(self.m_size_4)
        
        
        unique_4 = set(self.RoundingString(unique_4))

        self.m_unique_0 = [ ' '+item for item in unique_0 ]
        self.m_unique_1 = [ '  '+item for item in unique_1 ]
        self.m_unique_2 = [ '   '+item for item in unique_2 ]
        unique_3 = ['None'  if item == None else item for item in unique_3 ]
        self.m_unique_3 = [ '    '+item for item in unique_3 ]
        unique_4 = ['None'  if item == None else item for item in unique_4 ]
        self.m_unique_4 = [ '     '+item for item in unique_4 ]

        self.m_values_unique = self.m_unique_0 + self.m_unique_1 + self.m_unique_2 + self.m_unique_3 + self.m_unique_4
        return self.m_values_unique

    def Level(self):
        """
        Function that makes the column which has the level labels
        """
        self.m_level_treemap = []
        for i in range(len(self.m_unique_0)):
            self.m_level_treemap.append('Level 0')
        for i in range(len(self.m_unique_1)):
            self.m_level_treemap.append('Level 1')
        for i in range(len(self.m_unique_2)):
            self.m_level_treemap.append('Level 2')
        for i in range(len(self.m_unique_3)):
            self.m_level_treemap.append('Level 3')
        for i in range(len(self.m_unique_4)):
            self.m_level_treemap.append('Level 4')
        
        return self.m_level_treemap

    def Constant(self):
        """
        Function that adds the column with constant Random Forest value
        """
        
        #For a parent node add a constant string column
        self.m_constant = []
        for i in range(len(self.m_level_treemap)):
            self.m_constant.append('Random Forest')
        return self.m_constant
    
    def Size(self):
        self.ChildList()
        size_0 = []
        def counts(level_list):
            counts_dict = {}
            
            for item in level_list:
                if item in counts_dict:
                    counts_dict[item] = counts_dict[item] + 1
                else:
                    counts_dict[item] = 1
                counts_values = list(counts_dict.values())
            return counts_values

        size0 = counts(self.m_size_0)
        size1 = counts(self.m_size_1)
        size2 = counts(self.m_size_2)
        size3 = counts(self.m_size_3)
        size4 = counts(self.m_size_4)
        
        
        
        
        
        self.m_size = list(size0 + size1 + size2 + size3 + size4)
        
        return self.m_size
        
    def ConstructTreemapDF(self):
        """
        Function for constructing the dataframe
        """
        self.ChildList()
        self.Level()
        self.Constant()
        self.Size()
        
        df_treemap = pd.DataFrame()
        df_treemap.insert(loc = 0, column = 'Labels', value = self.m_level_treemap)
        df_treemap.insert(loc = 0, column = 'Values', value = self.m_values_unique)
        df_treemap.insert(loc = 0, column = 'Random', value = self.m_constant)
        df_treemap.insert(loc = 0, column = 'size', value = self.m_size)
        
        
        return df_treemap
```
